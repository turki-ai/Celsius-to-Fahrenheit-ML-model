```python
dic = {
    'input': [0, 8 , 15, 22],
    'output': [32, 46.4, 59, 71.6]
}
```

```python
import pandas as pd
df = pd.DataFrame(dic)
df
```

```python
x = df['input']
y= df['output']
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import warnings
warnings.filterwarnings('ignore')
```

```python

from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential

model = Sequential([
    Dense(1, input_shape=(1,), activation = 'linear')
])
```

```python
model.compile(
    loss = 'mse',
    metrics = ['mae'],
    optimizer = Adam()
)
```

```python
m = model.fit(
    x= x,
    y = y,
    epochs= 5000
)
```

```python
model.summary()
```

```python
print('loss:', m.history['loss'][-1])
```

```python
print('MAE:', m.history['mae'][-1])
```
